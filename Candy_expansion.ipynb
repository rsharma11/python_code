### Converting int columns to object

```python
int = ['int16', 'int32', 'int64']
others = ['object', 'float64']

all_int_col = candy_data.select_dtypes(include=int)
all_int_col = all_int_col.astype('str') 

other_col = candy_data.select_dtypes(include=others)

new_candy_data = pd.concat([all_int_col, other_col], axis=1)
new_candy_data.info()
```

### Exploring data insights

```python
# Univariate Plots (numerical data)
sns.boxplot(x="fruity", y="winpercent", data=new_candy_data).set_title('Relation of fruity and Win Percent');
```

```python
sns.boxplot(x="chocolate", y="winpercent", data=new_candy_data).set_title('Relation of Chocolate and Win Percent');
```

```python
sns.set(style="darkgrid", color_codes=True)
j = sns.jointplot('sugarpercent', 'winpercent', data = new_candy_data, kind='reg', height=8)
j.annotate(stats.pearsonr)
plt.show()
```

```python
sns.set(style="darkgrid", color_codes=True)
j = sns.jointplot('pricepercent', 'winpercent', data = new_candy_data, kind='reg', height=8)
j.annotate(stats.pearsonr)
plt.show()
```
